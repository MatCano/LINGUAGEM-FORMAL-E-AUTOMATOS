```python
# Criando jogo da velha e verificando o jogador usando o autômato criado

# Importando as bibliotecas necessárias para visualização gráfica
from graphviz import Digraph
from IPython.display import Image, display

# Definição da classe JogoDaVelhaDFA
class JogoDaVelhaDFA:
    def __init__(self):
        self.tabuleiro = [["v" for _ in range(3)] for _ in range(3)]  # Inicializa todas as casas com "v"
        self.jogador_atual = "x"
        self.historico_transicoes = []  # Armazena o histórico de transições

    # Método para imprimir o tabuleiro atual
    def imprimir_tabuleiro(self):
        print("  1 2 3")  # Imprime o cabeçalho das colunas
        # Itera sobre as linhas do tabuleiro
        for i, linha in enumerate(self.tabuleiro, start=1): # Imprime o número da linha e o conteúdo da linha separado por '|'
            print(f"{i} {'|'.join(linha)}")
            # Imprime a linha divisória entre as linhas do tabuleiro
            if i < 3:
                print("  -----")

    # Método para realizar uma jogada no tabuleiro
    def fazer_jogada(self, linha, coluna):
        linha -= 1
        coluna -= 1
        # Verifica se a casa escolhida está vazia
        if self.tabuleiro[linha][coluna] == "v":  # Verifica se a posição está disponível
            estado_atual = ''.join([''.join(linha) for linha in self.tabuleiro])    # Obtém o estado atual do tabuleiro como uma string
            self.tabuleiro[linha][coluna] = self.jogador_atual
            estado_resultante = ''.join([''.join(linha) for linha in self.tabuleiro])
            self.historico_transicoes.append((estado_atual, (linha+1, coluna+1, self.jogador_atual), estado_resultante))
            return True
        else:
            print("Posição ocupada, tente novamente.")
            return False

    def verificar_vitoria(self):

        # "Colorindo" estados finais para determinar quem ganhou

        # Estados finais que determinam vitória do x
        x = {'q4' , 'q14', 'q22', 'q31', 'q41', 'q51','q59' , 'q66'}
        verificar_x = NFA(q, sigma, delta, initial_state, x)

        # Determinando estados finais que determinam vitória do o
        o = {'q70' , 'q80' , 'q88' , 'q97', 'q107' , 'q117' , 'q125' , 'q132'}
        verificar_o = NFA(q, sigma, delta, initial_state, o)

        # Obtendo o estado atual do jogo e imprimindo como uma única string
        estado_atual = jogo.obter_estado_atual()
        print("Palavra atual do jogo:", estado_atual)

        # Verifica se o NFA aceita a palavra atual para o jogador 'x'
        if verificar_x.accept(estado_atual) == True:
            # Imprime o tabuleiro e anuncia que 'x' venceu
            self.imprimir_tabuleiro()
            print("x venceu!")
            return True
        # Verifica se o NFA aceita a palavra atual para o jogador 'o'
        elif verificar_o.accept(estado_atual) == True:
            # Imprime o tabuleiro e anuncia que 'o' venceu
            self.imprimir_tabuleiro()
            print("o venceu!")
            return True
        else:   # Se nenhum dos jogadores venceu, retorna False
            return False
    # Método para verificar se o jogo terminou em empate
    def verificar_empate(self):
        return all(cell != "v" for row in self.tabuleiro for cell in row)
    # Método para alternar entre os jogadores após cada jogada
    def alternar_jogador(self):
        # Alterna o jogador atual de 'x' para 'o', ou de 'o' para 'x'
        self.jogador_atual = "o" if self.jogador_atual == "x" else "x"

    def jogar(self):
        # Loop infinito para o jogo continuar até que haja um vencedor ou empate
        while True:
            # Imprime o estado atual do tabuleiro
            self.imprimir_tabuleiro()
            try:
                # Solicita ao jogador atual que escolha a linha e a coluna para a jogada
                linha = int(input(f"Vez do {self.jogador_atual}. Escolha a linha (1-3): "))
                coluna = int(input(f"Vez do {self.jogador_atual}. Escolha a coluna (1-3): "))
                # Verifica se os valores inseridos são válidos
                if linha < 1 or linha > 3 or coluna < 1 or coluna > 3:
                    print("Por favor, insira um número válido entre 1 e 3.")
                    continue
            except ValueError:
                # Captura erros se o jogador não inserir um número
                print("Por favor, insira um número válido entre 1 e 3.")
                continue
            # Realiza a jogada e verifica se foi bem-sucedida
            if self.fazer_jogada(linha, coluna):
                # Verifica se a jogada resultou em vitória
                if self.verificar_vitoria():
                    break
                # Verifica se todas as casas estão preenchidas e declara empate
                elif self.verificar_empate():
                    self.imprimir_tabuleiro()
                    print("Empatou!")
                    break
                # Se não houve vitória ou empate, alterna o jogador
                self.alternar_jogador()
    # Método para obter o estado atual do tabuleiro como uma string
    def obter_estado_atual(self):
        return ''.join([''.join(linha) for linha in self.tabuleiro])

# Rodar o jogo
jogo = JogoDaVelhaDFA()
jogo.jogar()
```
